# 카펫

## 문제
Leo는 카펫을 사러 갔다가 아래 그림과 같이 중앙에는 노란색으로 칠해져 있고 테두리 1줄은 갈색으로 칠해져 있는 격자 모양 카펫을 봄 <br>
Leo는 집으로 돌아와서 아까 본 카펫의 노란색과 갈색으로 색칠된 격자의 개수는 기억했지만, 전체 카펫의 크기는 기억하지 못함 <br>
Leo가 본 카펫에서 갈색 격자의 수 brown, 노란색 격자의 수 yellow가 매개변수로 주어질 때 카펫의 가로, 세로 크기를 순서대로 배열에 담아 return
<br><br>
input : brown = 정수, yellow = 정수, output : 정수 배열

#### <조건>
 - 갈색 격자의 수 brown은 8 이상 5,000 이하인 자연수
 - 노란색 격자의 수 yellow는 1 이상 2,000,000 이하인 자연수
 - 카펫의 가로 길이는 세로 길이와 같거나, 세로 길이보다 김
 
#### <예시>
input : brown = 10, yellow = 2
<br><br>
result : [4, 3]

## 풀이
 - 모든 카펫의 개수를 total에 저장
 - total로 만들 수 있는 모든 경우의 수를 구함
   - 이때, 가로가 세로보다 길거나 같기에 이 경우만 고려
   - 가운데 노랑색 타일이 있으려면 세로길이가 최소 2보다는 커야 함을 고려
 - yellow로 만들 수 있는 모든 경우의 수를 구함
   - 이떄ㅡ 가로가 세로보다 길거나 같기에 이 경우만 고려
   - yellow를 brown이 감싸고 있어야 하기에 yellow의 가로, 세로에 2씩 더해줌
 - yellow_div를 돌면서 total_div에 속하는 값이 있으면 return 

```python
def solution(brown, yellow):
    total = brown + yellow
    
    total_div = [[x, total//x] for x in range(total,0,-1) if total % x == 0 and total // x > 2 and x >= total // x]
    yellow_div = [[y+2, (yellow//y)+2] for y in range(yellow,0,-1) if yellow % y == 0 and y >= yellow //y]
    
    for y in yellow_div:
        if y in total_div:
            return y
```

```python
brown = 24
yellow = 24

solution(brown, yellow)
```
