## 문제
스파이들은 매일 다른 옷을 조합하여 입어 자신을 위장<br>
스파이가 가진 의상들이 담긴 2차원 배열 clothes가 주어질 때 서로 다른 옷의 조합의 수를 return
<br><br>
input : 2차원 문자열 배열, output : 정수

#### <조건>
 - clothes의 각 행은 [의상의 이름, 의상의 종류]로 이루어져 있음
 - 스파이가 가진 의상의 수는 1개 이상 30개 이하
 - 같은 이름을 가진 의상은 존재하지 않음
 - clothes의 모든 원소는 문자열로 이루어져 있음
 - 모든 문자열의 길이는 1 이상 20 이하인 자연수이고 알파벳 소문자 또는 '\_' 로만 이루짐
 - 스파이는 하루에 최소 한 개의 의상은 입음
 
#### <예시>
input : [["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]]
<br><br>
result : 5
<br><br>
 - headgear에 해당하는 의상이 yellow_hat, green_turban이고 eyewear에 해당하는 의상이 blue_sunglasses이므로 아래와 같이 5개의 조합이 가능
 
 - 따라서, 5를 return

## 풀이
### 1.
 - [의상 이름, 의상 종류] 순으로 주어지기에 의상 종류를 key값으로 딕셔너리 생성
   - 이때, 해당 의상이 선택이 안되는 경우가 존재하기에 ""빈 문자열도 추가로 넣어줌
 - 모든 의상을 다 넣어주고 나면 만들어진 clothes_dict을 돌면서 길이를 answer에 누적해서 곱해줌
 - 모든 의산이 선택되지 않을 경우인 1을 뺀 값을 return
 
### 2.
 - 1번 풀이를 모듈을 이용해서 간소화

```python
# 1.

def solution(clothes):
    clothes_dict = dict()
    
    for c in clothes:
        if c[1] in clothes_dict:
            clothes_dict[c[1]].append(c[0])
        else:
            clothes_dict[c[1]] = ["", c[0]]
    
    answer = 1
    for c in clothes_dict.values():
        answer *= len(c)
        
    return answer - 1
```

```python
# 2.

import collections
from functools import reduce

def solution(clothes):
    clothes_dict = collections.Counter(c[1] for c in clothes)
    
    return reduce(lambda std, c: std*(c+1), clothes_dict.values(), 1) - 1
```

```python
clothes = [["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]]

solution(clothes)
```
