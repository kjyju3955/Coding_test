## 이중우선순위큐

## 문제
이중 우선순위 큐는 다음 연산을 할 수 있는 자료구조



이중 우선순위 큐가 할 연산 operations가 매개변수로 주어질 때, 모든 연산을 처리한 후 큐가 비어있으면 [0,0] 비어있지 않으면 [최댓값, 최솟값]을 return
<br><br>
input : 문자열배열, output : 정수 배열

#### <조건>
 - operations는 길이가 1 이상 1,000,000 이하인 문자열 배열
 - operations의 원소는 큐가 수행할 연산을 나타냄
 - 원소는 “명령어 데이터” 형식으로 주어짐
   - 최댓값/최솟값을 삭제하는 연산에서 최댓값/최솟값이 둘 이상인 경우, 하나만 삭제
 - 빈 큐에 데이터를 삭제하라는 연산이 주어질 경우, 해당 연산은 무시
 
### <예시>
input : ["I 7","I 5","I -5","D -1"]
<br><br>
result : [7, 5]
<br><br>
 - 7,5,-5를 삽입 
 - 최소값 삭제
 - 최대값 7, 최소값 5를 반환

## 풀이
 - 연산에 결과를 담을 answer이라는 빈 리스트 선언
 - operations를 앞에서부터 하나씩 처리
 - 반복을 돌 때마다 answer를 정렬
 - operation에서 I일 경우에는 뒤에오는 숫자를 answer에 넣어줌
 - D -1 일 경우에는 answer는 정렬되어 있는 상태이기에 첫 번째 숫자를 제거
   - 이때, answr에 숫자가 없는 경우에는 동작하지 않아야 함
 - D 1일 경우에는 answer의 마지막 숫자를 제거
 - 모든 operation을 돌고 난 후에 answer가 존재하면 다시 정렬 해주고 제일 마지막 값과 제일 첫 번째 값을 return
 - answer가 존재하지 않으면 0, 0을 return

```python
def solution(operations):
    answer = []
    
    while operations:
        std = operations.pop(0)
        
        answer = sorted(answer)
        
        if std[0] == 'I':
            std = std.split(' ')
            answer.append(int(std[1]))
        elif answer and std == 'D -1':
            answer.pop(0)
        elif answer and std == 'D 1':
            answer.pop()
            
    if answer:
        answer = sorted(answer)
        return [answer[-1], answer[0]]
    
    return [0, 0]
```

```python
operations = ["I 4", "I 3", "I 2", "I 1", "D 1", "D 1", "D -1", "D -1", "I 5", "I 6"] 

solution(operations)
```
