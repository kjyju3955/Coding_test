# 2 x n 타일링

## 문제
가로 길이가 2이고 세로의 길이가 1인 직사각형모양의 타일이 있음. 이 직사각형 타일을 이용하여 세로의 길이가 2이고 가로의 길이가 n인 바닥을 가득 채우려고 함. 타일을 채울 때는 다음과 같이 2가지 방법이 있음.
 - 타일을 가로로 배치 하는 경우
 - 타일을 세로로 배치 하는 경우 
직사각형의 가로의 길이 n이 매개변수로 주어질 때, 이 직사각형을 채우는 방법의 수를 return.

<br><br>
input : 정수, output : 정수

#### <조건>
 - 가로의 길이 n은 60,000이하의 자연수
 - 경우의 수가 많아 질 수 있으므로, 경우의 수를 1,000,000,007으로 나눈 나머지를 return
 
#### <예시>
input : 4
<br><br>
result : 5
<br><br>
 - 다음과 같이 5개의 방법이 존재

## 풀이
 - n이 1일 때부터 가지수를 계산해보면 1, 2, 3, 5, 8, .... 순으로 피보나치 수열의 모습을 보인다
   - 어쩌다보니 힌트를 봐버림...ㅠ
 - 피보나치 수열은 앞의 두 수를 더해서 다음수를 만들어주는 수열이기에 두 수를 저장해줄 num1, num2 선언
 - 두 수를 제외하고 3번째부터 n번째까지 계속해서 계산을 해줌
 - n번쨰까지 계산을 해주고 난 후에 요구조건에 있던 1000000007로 나눈 나머지를 return

```python
def solution(n):
    num1, num2 = 1, 2
    
    for i in range(3, n+1):
        tmp = num2
        num2 = num1 + num2
        num1 = tmp
    
    return num2 % 1000000007
```

```python
n = 4

solution(n)
```
