# 괄호 회전하기

## 문제
다음 규칙을 지키는 문자열을 올바른 괄호 문자열이라고 정의
 - (), [], {} 는 모두 올바른 괄호 문자열
 - 만약 A가 올바른 괄호 문자열이라면, (A), [A], {A} 도 올바른 괄호 문자열 <br>
   예를 들어, [] 가 올바른 괄호 문자열이므로, ([]) 도 올바른 괄호 문자열
 - 만약 A, B가 올바른 괄호 문자열이라면, AB 도 올바른 괄호 문자열 <br>
   예를 들어, {} 와 ([]) 가 올바른 괄호 문자열이므로, {}([]) 도 올바른 괄호 문자열
대괄호, 중괄호, 그리고 소괄호로 이루어진 문자열 s가 매개변수로 주어짐. 이 s를 왼쪽으로 x (0 ≤ x < (s의 길이)) 칸만큼 회전시켰을 때  <br>
s가 올바른 괄호 문자열이 되게 하는 x의 개수를 return 
<br><br>
input : 문자열, output : 정수

#### <조건>
 - s의 길이는 1 이상 1,000 이하

#### <예시>
 
 
 - 올바른 괄호 문자열이 되는 x가 3개이므로, 3을 return

## 풀이
 - 문자열을 리스트로 변경
 - 올바른 괄호일경우는 1, 아닐 경우는 0을 저장해주기 위해서 0으로 초기화한 리스트 answer 선언
 - s를 왼쪽으로 i칸만큼 회전시키면서 올바른 괄호인지 확인하는 check_list에 검사
   - stack 활용
   - s를 앞에서부터 하나씩 pop해서 [,\{,\(인 경우에는 stack에 넣어주고 아닐 경우에는 따로 처리를 해줌
   - 아닐 경우 stack이 비어있으면 올바른 괄호가 아니기에 0을 return 해줌
   - 아스키코드 값을 이용해서 stack이 비어있지 않고 짝이 아닐 시에는 0을 return
   - s를 다돌고 난 후에도 stack에 무언가가 남아있다면 짝이 없는 것이기에 0을 return
   - 그외의 경우는 올바른 괄호이기에 1을 return
 - return 한 값들을 answer에저장해주고 1의 개수를 count해서 return

```python
def solution(s):
    s = list(s)
    answer = [0 for i in s]

    for i in range(len(s)):
        answer[i] = check_list(s.copy())
        s.append(s.pop(0))
        
    return answer.count(1)
        
def check_list(s):
    c_stack = []
    
    while s:
        std = s.pop(0)
        
        if std == '[' or std == '{' or std == '(':
            c_stack.append(std)
            continue
        
        if not c_stack:
            return 0
        
        if abs(ord(std) - ord(c_stack[-1])) > 2:
            return 0
        
        c_stack.pop()
        
    if c_stack:
        return 0
            
    return 1
```

```python
s = "{"

solution(s)
```
